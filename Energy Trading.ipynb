Finding the % change per day using previous day/lagging

```python
ret = close.pct_change().dropna()

# Customize display options for better output
pd.set_option('display.float_format', '{:.4%}'.format)
pd.set_option('display.max_rows', 10)

# Print the first few rows of the DataFrame
print(ret.head())
```

```python
#Average Daily Return and Standard Deviation
summary = ret.describe().T.loc[:,['mean','std']]
summary
```

```python
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
summary
```

```python
# Get the list of stock names from retun data
columns = ret.columns

# Set up subplots for each column
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(18, 12))

# Flatten the 2D array of subplots to a 1D array
axes = axes.flatten()

# Iterate over columns and plot histograms
for i, column in enumerate(columns):
    ret[column].plot(kind='hist', bins=100, ax=axes[i])
    axes[i].set_title(column)
    
    # adjust the x-axis limits for each subplot to zoom in
    axes[i].set_xlim([min(ret[column]), max(ret[column])])

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

```python
summary.plot.scatter(x ='std', y ='mean', figsize =(12,8), s=50, fontsize=10)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+.002,summary.loc[i,"mean"]+.002),size=9)
plt.xlabel("Annual Risk(std)", fontsize = 10)
plt.ylabel("Annual Risk(mean)", fontsize = 10)
plt.title(("Risk Return"), fontsize = 12)
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(ret.corr(),cmap="Reds",annot=True,annot_kws={"size":12},vmax=0.6)
plt.show()
```
