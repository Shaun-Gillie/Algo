Import necessary packages

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ticker = ["XOM","CVX","SHEL", "COP", "TTE", "BP", "ENB"]
```

```python
stocks = yf.download(ticker, start = "2010-01-01", end = "2024-01-01")
stocks
```

```python
stocks.info()
```

Export Selected Energy Stocks as CSV and import and Pandas Dataframes

```python
stocks.to_csv("stocksYT.csv")
```

```python
stocks = pd.read_csv("stocksYT.csv", header = [0,1], index_col=[0], parse_dates=[0])
stocks
```

```python
#convert multi index into one tuple
stocks.columns = stocks.columns.to_flat_index()
stocks.columns
```
